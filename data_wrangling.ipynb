```python
revised_data
```

```python
country_metadata = pd.read_csv(r"World Debt Analysis/Dataset/IDS_CountryMetaData.csv",encoding="ANSI")
#Reading the metadata file
```

```python
country_metadata.sample(n=5)
```

```python
income_groups = country_metadata["Income Group"].unique().tolist()
#collecting unique income groups
```

```python
regions = country_metadata["Region"].unique().tolist()
#collecting unique regions
```

```python
lst =[]
```

```python
for item in income_groups:
    lst.append(revised_data.index[revised_data.country_name == item].tolist())
#Appending income groups to a new list
```

```python
len(lst)
```

```python
for item in regions:
    lst.append(revised_data.index[revised_data.country_name == item].tolist())
#Appending regions to the same list with income groups already in it.
```

```python
len(lst)
```

```python
new_lst = list(chain.from_iterable(lst))
#Unpacking the previous list because it was nested list and creating a 1D list.
```

```python
len(new_lst)
```

```python
revised_data.drop(new_lst, inplace = True)
#Dropping not valid country names from the data
```
